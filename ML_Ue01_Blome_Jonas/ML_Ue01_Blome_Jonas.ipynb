```python
# 6. Create the appropriate linear equation and store it in a function.

nations = np.genfromtxt("nations.csv", delimiter=",", skip_header=True)

import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel("GDP per capita")
plt.ylabel("Average life expectancy")

gdppc = nations[:,3]

for i in range(len(gdppc)):
    gdppc[i] = gdppc[i] / nations[i, 6]

plt.plot(gdppc, nations[:,4], "ob")
plt.plot([0, 0.14], [71, 80])

def linearEquation(x):
    return 64.3 * x + 71

plt.plot([0, 0.14], [linearEquation(0), linearEquation(0.14)])
```

```python
# 7. Calculate how well the line approximates the data points by calculating the root-mean-square error (RMSE).

nations = np.genfromtxt("nations.csv", delimiter=",", skip_header=True)

import matplotlib.pyplot as plt
%matplotlib inline
import math

plt.xlabel("GDP per capita")
plt.ylabel("Average life expectancy")

gdppc = nations[:,3]

for i in range(len(gdppc)):
    gdppc[i] = gdppc[i] / nations[i, 6]

plt.plot(gdppc, nations[:,4], "ob")

def linearEquation(x):
    return 64.3 * x + 71

squareErrorSum = 0

for i in range(len(gdppc)):
    squareErrorSum = squareErrorSum + pow(linearEquation(gdppc[i]) - nations[i, 4], 2)

mse = squareErrorSum / len(nations[:,4])

rmse = math.sqrt(mse)

print(rmse)

plt.plot([0, 0.14], [linearEquation(0), linearEquation(0.14)])
```
